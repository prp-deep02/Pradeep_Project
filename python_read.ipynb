```python
from google.colab import drive
import pandas as pd
drive.mount('/content/drive')
```

```python
empcsv=pd.read_csv('/content/drive/MyDrive/employeess.csv')
emp=pd.DataFrame(empcsv)
print(emp)
```

```python
print(emp.isnull().sum())
```

```python
emp['Age']=emp['Age'].fillna(emp['Age'].median())
print(emp['Age'].isnull().sum())
emp['Salary']=emp['Salary'].fillna(emp['Salary'].median())
print(emp['Salary'].isnull().sum())
```

```python
emp['Department']=emp['Department'].str.capitalize()
print(emp['Department'].head())
```

```python
print(emp.drop_duplicates())
```

```python
emp.to_csv("cleaned_employees.csv")
```
